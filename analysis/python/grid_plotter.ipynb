```python
import numpy as np
import sklearn
import matplotlib
import pandas as pd
import os
import matplotlib
import matplotlib.pyplot as plt 
```

```python
# lab = pd.read_csv('../../data/UW58lab.csv', header=None)
rgb = pd.read_csv('../../data/UW_58_rgb.csv',header= None)

cols =  pd.read_csv('../../data/UW58_colors.csv')
```

```python
lab.columns = ['L','a','b']
rgb.columns = ['R','G','B']
plot_df = pd.concat([lab,rgb], axis=1)
```

```python
plot_df = pd.concat([cols,rgb], axis=1)
```

```python
plot_df.rename(columns=dict(zip([0,1,2],['R','G','B'])), inplace=True)
```

```python
plt.figure(figsize=(10,17))
plt.scatter(plot_df.a, plot_df.b+10*plot_df.L, s=2500, marker= (4, 0, 45), color = list(zip(plot_df.R,plot_df.G,plot_df.B)))
ax = plt.gca()
ax.set_facecolor((0.3492,0.3492,0.3492))
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)
#plt.savefig('patches71.pdf',bbox_inches='tight',pad_inches=0)
```
